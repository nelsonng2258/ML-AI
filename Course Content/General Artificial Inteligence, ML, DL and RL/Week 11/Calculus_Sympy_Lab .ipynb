## Limits

```python
limit(sin(x)/x, x, 0)
```

To evaluate a limit at one side only, pass '+' or '-' as a third argument to limit.

```python
#approach from the right
expr = limit(1/x, x, 0, '+')
expr
```

```python
#approach from the left
expr = limit(1/x, x, 0, '-')
expr
```

## Series Expansion
SymPy can compute asymptotic series expansions of functions around a point. To compute the expansion of f(x) around the point x=x0 terms of order x^n , use f(x).series(x, x0, n). x0 and n can be omitted, in which case the defaults x0=0 and n=6 will be used.

```python
f = (x**2)*y

f.series(x, x0=3,n = 2) + f.series(y, x0=8,n = 2)
```

```python
from sympy.abc import x, y, z, n, a

f = (x/(x**2+y**2))

f.series(x, x0=1,n = 2) + f.series(y, x0=2,n = 2)
```

```python
expr = exp(sin(x))
expr.series(x, 0, 4)
expr.series(x)
```

```python
import math

expr = cos(2*x)
expr.series(x, sympy.pi, 6)
```

```python
expr = 1+x+x**2
expr.series(x0=1, n= 4)
```

```python
expr = (x+10)**(1/2)

expr.series(x0=1, n= 4)
```

Ff you do not want the order term, use the removeO method.

```python
expr.series(x, 0, 4).removeO()
```

```python
exp(x - 6).series(x, x0=6)
```
